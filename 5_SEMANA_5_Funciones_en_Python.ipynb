### Celda 2: Texto (Markdown)

```markdown
## SESIÓN 1: Introducción a las Funciones

**El Problema de la Repetición**
Supongamos que en un levantamiento topográfico necesitamos calcular el área de 50 terrenos rectangulares distintos. Escribir la operación `base * altura` 50 veces haría el programa largo y difícil de mantener.

**La Solución: Funciones**
Una función es un bloque de código encapsulado que realiza una tarea específica y que podemos "llamar" o ejecutar todas las veces que queramos.

### 1. Definición y Llamado de una Función Simple
Para crear una función en Python usamos la palabra clave `def`, seguida del nombre que queramos darle y paréntesis `()`. Todo lo que esté indentado (con un espacio hacia la derecha) pertenecerá a esa función.

```

### Celda 3: Código (Python)

````
# Definición de la función (El programa la aprende, pero aún no hace nada visible)
def mostrar_alerta_climatica():
    print("ALERTA: Se pronostican fuertes lluvias en la región de la sabana.")
    print("Revisar niveles de drenaje en los cultivos.")

# Llamado de la función (Aquí es donde realmente se ejecuta)
mostrar_alerta_climatica()
````

### Celda 4: Texto (Markdown)

```markdown
### 2. Funciones con Parámetros (Entradas)
Las funciones son mucho más útiles cuando pueden recibir datos para procesarlos. A estos datos los llamamos **parámetros**. Esto coincide perfectamente con nuestro modelo algorítmico de **Entrada $\rightarrow$ Proceso $\rightarrow$ Salida**.

```

### Celda 5: Código (Python)

````
# Función que recibe dos parámetros: base y altura
def calcular_area_terreno(base, altura):
    area = base * altura
    print("El área calculada es:", area, "metros cuadrados.")

# Reutilizamos la misma función con diferentes datos
print("Terreno 1:")
calcular_area_terreno(15.5, 20.0)

print("\nTerreno 2:")
calcular_area_terreno(8.0, 12.0)
````

```markdown
## SESIÓN 2: Funciones con Retorno de Valores (`return`)

Hasta ahora, nuestras funciones calculan algo y lo imprimen directamente en pantalla con `print()`. Sin embargo, en la ingeniería profesional, usualmente necesitamos que la función **nos devuelva el valor matemático** para usarlo en otras fórmulas más adelante.

Para esto usamos la palabra clave `return`.

### Ejemplo Aplicado: Volumen de un Tanque de Almacenamiento
Imagina que estás diseñando un sistema de riego o acueducto y necesitas calcular la capacidad de un tanque cilíndrico. La fórmula es $V = \pi r^2 h$.

```

### Celda 8: Código (Python)

````
import math

# 1. Definimos la función que hace el cálculo y RETORNA el valor
def volumen_cilindro(radio, altura):
    volumen = math.pi * (radio**2) * altura
    return volumen  # Entrega el dato de vuelta al programa principal

# 2. Pedimos los datos al usuario (Entradas)
r = float(input("Ingrese el radio del tanque en metros: "))
h = float(input("Ingrese la altura del tanque en metros: "))

# 3. Llamamos a la función y guardamos su resultado en una nueva variable 'v'
v = volumen_cilindro(r, h)

# 4. Mostramos el resultado final
print("El volumen útil del tanque es de:", v, "metros cúbicos.")
````

### Celda 9: Texto (Markdown)

```markdown
### Mini-proyecto: Cálculo de Velocidad Media
En topografía o diseño de canales, calcular velocidades es rutinario ($v = \frac{d}{t}$). Ejecuta y analiza este código modular.

```

### Celda 10: Código (Python)

````
def velocidad_media(distancia, tiempo):
    v = distancia / tiempo
    return v

# Bloque principal del programa
d = float(input("Ingrese la distancia recorrida (m): "))
t = float(input("Ingrese el tiempo transcurrido (s): "))

vel_calculada = velocidad_media(d, t)
print("La velocidad media del flujo o vehículo es:", vel_calculada, "m/s")
````
