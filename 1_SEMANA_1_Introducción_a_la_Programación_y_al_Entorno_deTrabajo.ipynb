```markdown
## SESIÓN 1: Introducción al Pensamiento Algorítmico

**¿Qué es programar?**
Programar consiste en describir, paso a paso, cómo resolver un problema mediante instrucciones lógicas que una computadora puede ejecutar.

Para estructurar estas soluciones utilizamos el **Modelo E-P-S (Entradas - Proceso - Salidas)**. Veamos cómo se aplica esto en sus áreas de estudio:

**1. Ingeniería Civil:**
* **Problema:** Calcular el volumen de concreto necesario para una zapata rectangular.
* **Entradas:** largo, ancho, profundidad.
* **Proceso:** multiplicar largo * ancho * profundidad.
* **Salida:** volumen de concreto en metros cúbicos.

**2. Ingeniería Agrícola:**
* **Problema:** Calcular el caudal de un canal rectangular.
* **Entradas:** velocidad del flujo, área de la sección transversal.
* **Proceso:** multiplicar la velocidad por el área.
* **Salida:** caudal en metros cúbicos por segundo.

**3. Ingeniería Agroindustrial:**
* **Problema:** Determinar el índice de madurez de una fruta.
* **Entradas:** grados Brix (azúcares), porcentaje de acidez.
* **Proceso:** dividir los grados Brix entre el porcentaje de acidez.
* **Salida:** índice de madurez.
```

```markdown
## SESIÓN 2: Aplicación Práctica y Síntesis Conceptual

Vamos a aplicar el modelo Entradas-Proceso-Salidas en un caso práctico. Analiza el siguiente código, el cual automatiza el cálculo del área de un terreno, una tarea común en topografía y diseño agrícola.

```

### Celda 9: Código (Python)

````
# Programa para calcular el área de un terreno rectangular

# 1. ENTRADAS (Pedimos los datos al usuario y los convertimos a números decimales con 'float')
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# 2. PROCESO (Realizamos el cálculo matemático)
area = largo * ancho

# 3. SALIDA (Mostramos el resultado final)
print("El área total del terreno es de:", area, "metros cuadrados.")
````
