```markdown
## SESIÓN 1: El Eslabón Perdido - El archivo `requirements.txt`

Antes de subir nuestra app a la nube, necesitamos explicarle a los servidores de Streamlit qué herramientas externas (librerías) utilizamos en nuestro código.

Si en nuestro archivo `app.py` escribimos `import streamlit` o `import math`, el servidor necesita saber que debe instalar esas piezas antes de correr el programa. Esto se hace mediante un archivo de texto simple.

**Paso a paso en VSCode:**
1. Abre tu proyecto en Visual Studio Code.
2. En la misma carpeta donde está tu `app.py`, crea un nuevo archivo y nómbralo exactamente así: **`requirements.txt`**
3. Abre ese archivo y escribe los nombres de las librerías externas que usaste, una por línea.

```

```markdown
### Actualizando nuestro repositorio en GitHub
Como agregamos un archivo nuevo (`requirements.txt`), debemos enviarlo a la nube. Abre la terminal en VSCode y ejecuta la secuencia que aprendimos la semana pasada:

1. `git add .` (Para preparar el nuevo archivo)
2. `git commit -m "Agregado archivo de requerimientos"` (Para tomar la fotografía del cambio)
3. `git push` (Para enviar los cambios a GitHub)

```

```markdown
## SESIÓN 2: El Despliegue en Streamlit Community Cloud

Streamlit nos ofrece un servicio gratuito llamado **Community Cloud** que se conecta directamente a nuestro GitHub y publica nuestras aplicaciones al instante.

**Paso a paso para el Despliegue:**
1. Ve a tu navegador e ingresa a [share.streamlit.io](https://share.streamlit.io/).
2. Haz clic en **"Continue with GitHub"** e inicia sesión con tu cuenta.
3. Autoriza a Streamlit para que pueda leer tus repositorios.
4. Una vez dentro de la plataforma, haz clic en el botón azul **"New app"** (Nueva aplicación).
5. Se abrirá un formulario sencillo. Llena los datos seleccionando desde las listas desplegables:
   * **Repository:** Busca y selecciona tu repositorio (por ejemplo, si tu usuario es *justorfc*, selecciona `justorfc/calculadora-tanques`).
   * **Branch:** Selecciona `main` (o `master`).
   * **Main file path:** Escribe `app.py` (el nombre de tu archivo principal).
   * **App URL:** Aquí puedes personalizar el enlace que tendrá tu aplicación para que sea fácil de recordar.
6. Haz clic en el botón **"Deploy!"** y espera unos momentos. Verás en la pantalla cómo el servidor instala tus requerimientos y lanza tu aplicación.

¡Felicidades, ingeniero! Tu aplicación ya está viva en Internet. Copia la URL de tu navegador y envíasela a un compañero para que la pruebe desde su celular.

```

### Celda 6: Texto (Markdown)

```markdown
### La Magia de la Integración Continua (CI/CD)

Lo mejor de haber conectado Streamlit con GitHub es que, a partir de ahora, **cualquier cambio que hagas en tu computador se actualizará automáticamente en la web**.

Haz la prueba:
1. Ve a VSCode y cambia el título de tu aplicación (por ejemplo, agrega un emoji nuevo).
2. Guarda el archivo.
3. Usa `git add .`, `git commit -m "Cambio de título"` y `git push` en tu terminal.
4. Ve a la URL pública de tu aplicación en el navegador y verás cómo, tras unos segundos, el título cambia automáticamente. ¡Es magia pura!

```
