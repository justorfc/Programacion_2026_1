## SESIÓN 1: Introducción a las Decisiones en Programación

**¿Qué ocurre si un programa debe tomar decisiones?**
En la ingeniería del mundo real, las acciones dependen de las mediciones. Por ejemplo:
* *Ingeniería Agrícola:* Si la humedad del suelo es menor a un valor crítico, se debe activar el riego. Si no, se mantiene apagado.
* *Ingeniería Civil:* Si la carga sobre una viga supera el límite de fluencia, se emite una alerta estructural.
* *Ingeniería Agroindustrial:* Si la temperatura de un cuarto frío sube de 4°C, se debe encender el compresor de respaldo.

**Modelo Lógico Fundamental:**
> **Si** (la condición es verdadera):
> &nbsp;&nbsp;&nbsp;&nbsp;ejecutar acción A
> **Si no**:
> &nbsp;&nbsp;&nbsp;&nbsp;ejecutar acción B

### Operadores Relacionales
Las decisiones se construyen comparando valores. En Python utilizamos los siguientes operadores:
* `>` (Mayor que)
* `<` (Menor que)
* `>=` (Mayor o igual)
* `<=` (Menor o igual)
* `==` (Igual a - *¡Nota que son dos signos de igual!*)
* `!=` (Diferente de)

### Celda 3: Texto (Markdown)

### Primera Estructura Condicional: El uso de `if`
Evaluemos un proceso agroindustrial. Si la temperatura pasa de cierto límite, mostraremos una alerta. Ejecuta la celda e ingresa un valor como 85, y luego vuelve a ejecutarla con un valor de 70.

### Celda 4: Código (Python)

````
# Ejemplo 1: Condicional simple (if)
temperatura = float(input("Ingrese temperatura del proceso (°C): "))

if temperatura > 80:
    print("¡Alerta! Temperatura alta detectada en el sistema.")
````

### Celda 5: Texto (Markdown)

```markdown
### Estructura `if - else` (Si - Si no)
Apliemos esto a un sistema de riego (Ingeniería Agrícola). Aquí el programa decide entre dos caminos posibles.

```

### Celda 6: Código (Python)

````
# Ejemplo 2: Condicional doble (if - else)
humedad = float(input("Ingrese la humedad actual del suelo (%): "))

if humedad < 30:
    print("Acción: Activar sistema de riego.")
else:
    print("Acción: No es necesario regar. Humedad óptima.")
````

### Celda 7: Texto (Markdown)

```markdown
### Múltiples Condiciones: `if - elif - else`
A veces hay más de dos opciones. Por ejemplo, clasificar el pH del suelo requiere evaluar varios rangos.

```

### Celda 8: Código (Python)

````
# Ejemplo 3: Condicional múltiple (if - elif - else)
ph = float(input("Ingrese el valor de pH del suelo: "))

if ph < 5.5:
    print("Clasificación: Suelo ácido")
elif ph <= 7.0:
    print("Clasificación: Suelo neutro")
else:
    print("Clasificación: Suelo alcalino")
````

```markdown
## SESIÓN 2: Aplicación Práctica de Estructuras Condicionales

Vamos a construir un programa integral para clasificar la **eficiencia de un sistema** (muy útil en evaluación de maquinaria o procesos de producción).

**Reglas del negocio:**
* Eficiencia mayor o igual a 90% $\rightarrow$ Alta
* Eficiencia entre 70% y 89.9% $\rightarrow$ Media
* Eficiencia menor a 70% $\rightarrow$ Baja

```

### Celda 11: Código (Python)

````
# Ejercicio Práctico: Clasificación de Eficiencia

# 1. ENTRADA
eficiencia = float(input("Ingrese la eficiencia del sistema (%): "))

# 2. PROCESO Y 3. SALIDA (Evaluación de condiciones)
if eficiencia >= 90:
    print("Diagnóstico: Eficiencia alta")
elif eficiencia >= 70:
    print("Diagnóstico: Eficiencia media")
else:
    print("Diagnóstico: Eficiencia baja")
````
