### Solución 1: Ingeniería Agroindustrial

**Tema:** Conversión de tipos de datos y cálculo de porcentajes.

```
# EJERCICIO: CONTROL DE RENDIMIENTO DE ACEITE (AGROINDUSTRIAL)
# -----------------------------------------------------------

# 1. Entrada de datos (Recordar que input() devuelve texto/str)
nombre_lote = input("Ingrese el nombre o código del lote: ")
peso_rff = float(input("Ingrese el peso de los racimos (kg): "))
litros_aceite = float(input("Ingrese los litros de aceite obtenidos: "))

# 2. Definición de constantes y procesos
DENSIDAD_ACEITE = 0.92  # kg/L

# Cálculo del peso en kg del aceite obtenido
peso_aceite_kg = litros_aceite * DENSIDAD_ACEITE

# Cálculo del porcentaje de rendimiento
# Fórmula: (Peso Final / Peso Inicial) * 100
rendimiento = (peso_aceite_kg / peso_rff) * 100

# 3. Salida de resultados
print("-" * 30)
print(f"REPORTE DEL LOTE: {nombre_lote}")
print(f"Peso del aceite extraído: {peso_aceite_kg} kg")
print(f"El rendimiento final es del: {rendimiento}%")
```

### Solución 2: Ingeniería Civil

**Tema:** Uso de librerías (`math`), potencias y redondeo.

```
# EJERCICIO: DOSIFICACIÓN DE CEMENTO EN COLUMNAS (CIVIL)
# -----------------------------------------------------
import math # Importamos para usar el valor preciso de PI

# 1. Entrada de datos
altura = float(input("Ingrese la altura de la columna (m): "))
diametro = float(input("Ingrese el diámetro de la columna (m): "))

# 2. Procesos Geométricos y Técnicos
radio = diametro / 2

# Cálculo del volumen: PI * R^2 * H
# Usamos math.pi para precisión y el operador ** para la potencia
volumen_m3 = math.pi * (radio ** 2) * altura

# Según norma: 9.5 sacos de cemento por cada 1 m3 de concreto 210 kg/cm2
SACOS_POR_M3 = 9.5
total_sacos = volumen_m3 * SACOS_POR_M3

# 3. Salida de resultados
print("-" * 30)
print(f"RESULTADOS DEL CÁLCULO:")
print(f"Volumen de la columna: {volumen_m3:.4f} metros cúbicos")
print(f"Cantidad de cemento necesaria: {total_sacos:.2f} sacos")
```
