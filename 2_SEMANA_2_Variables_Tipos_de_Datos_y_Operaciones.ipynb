```markdown
## SESIÓN 1: Variables y Tipos de Datos

**¿Qué es una variable?**
Una variable es un espacio reservado en la memoria del computador donde almacenamos un dato que nuestro programa necesita. Por ejemplo, si medimos la temperatura con un sensor, podemos guardarla así: `temperatura = 28.5`.

En sus áreas de ingeniería, las variables representan magnitudes físicas y datos críticos:
* **caudal:** flujo de agua en un distrito de riego (Agrícola).
* **temperatura:** medición térmica en un proceso de secado (Agroindustrial).
* **presion:** carga soportada en el diseño de una estructura (Civil).

Existen tres tipos de datos básicos que usaremos hoy:
1. **Enteros (`int`):** Números sin decimales (ej. `equipos = 5`).
2. **Decimales (`float`):** Números con precisión decimal (ej. `temperatura = 27.5`).
3. **Texto (`str`):** Cadenas de caracteres (ej. `material = "Concreto"`).

```

### Celda 3: Código (Python)

````
# Creación de variables con diferentes tipos de datos
caudal_canal = 15          # int (entero)
temperatura_horno = 185.5  # float (decimal)
material_construccion = "Concreto Reforzado" # str (texto)

print("El material evaluado es:", material_construccion)
print("La temperatura del horno es:", temperatura_horno, "grados")
````

```markdown
### Operadores y Precedencia

Para transformar estos datos, utilizamos operadores aritméticos:
* Suma (`+`), Resta (`-`), Multiplicación (`*`), División (`/`).
* Potencia (`**`) y Residuo (`%`).

**Precedencia de operaciones:**
Python resuelve las matemáticas con las mismas reglas fundamentales:
1. Paréntesis `()`
2. Potencias `**`
3. Multiplicación `*` y División `/`
4. Suma `+` y Resta `-`

*Ejemplo:* En la expresión `5 + 3 * 2`, Python primero multiplica (3*2=6) y luego suma 5, dando como resultado `11`.

```

### Celda 5: Código (Python)

````
# Ejemplo de operadores: Cálculo de producción promedio
produccion_total = 120
dias_trabajados = 6

promedio = produccion_total / dias_trabajados
print("La producción promedio diaria es:", promedio)

# Demostración de precedencia
resultado_1 = 5 + 3 * 2
resultado_2 = (5 + 3) * 2

print("Resultado sin paréntesis:", resultado_1)
print("Resultado con paréntesis:", resultado_2)
````

```markdown
## SESIÓN 2: Aplicación Práctica con Variables

Vamos a aplicar esto en un programa que calcula la **eficiencia de producción**. Este cálculo es fundamental para evaluar el rendimiento de una máquina cosechadora, una planta de extracción o el avance de una cuadrilla de obra.
```

### Celda 8: Código (Python)

````
# Cálculo de eficiencia de producción

# 1. ENTRADAS (Pedimos datos y los convertimos a decimales)
produccion_real = float(input("Ingrese la producción real obtenida: "))
produccion_esperada = float(input("Ingrese la meta de producción esperada: "))

# 2. PROCESO (Uso de paréntesis para asegurar la precedencia correcta)
eficiencia = (produccion_real / produccion_esperada) * 100

# 3. SALIDA
print("La eficiencia del sistema es:", eficiencia, "%")
````
