**¿Qué es Streamlit?**
Es una herramienta de Python que permite crear aplicaciones web para ciencia de datos e ingeniería en cuestión de minutos, sin necesidad de saber diseño web (HTML, CSS o JavaScript).

**El Mapeo de Conceptos (De la consola a la web):**
* Antes pedíamos datos así: `radio = float(input("Ingrese radio: "))`
* Ahora lo haremos con cajas numéricas: `radio = st.number_input("Ingrese radio: ")`
* Antes mostrábamos resultados así: `print("El volumen es", v)`
* Ahora lo haremos con textos elegantes: `st.success(f"El volumen es {v}")`

### Escribiendo nuestra Primera Aplicación Web

Vamos a tomar la función de **Cálculo de Volumen de un Tanque** que hicimos la semana pasada (muy útil para almacenamiento agrícola o diseño civil) y la convertiremos en una App.

En Colab, usamos el comando `%%writefile app.py` en la primera línea para indicarle al entorno que no ejecute el código aquí mismo, sino que lo guarde en un archivo llamado **app.py**. Ejecuta la celda:

### Celda 5: Código (Python)

````
%%writefile app.py
import streamlit as st
import math

# 1. Nuestra función lógica (El cerebro de la app)
def volumen_cilindro(radio, altura):
    return math.pi * (radio**2) * altura

# 2. Diseño de la Interfaz Web
st.title("Calculadora de Volumen para Tanques 💧")
st.write("Aplicación interactiva para Ingeniería Agrícola y Civil.")
st.divider() # Dibuja una línea separadora

# 3. Entradas de datos (Reemplazan a los inputs)
st.subheader("Ingrese las dimensiones:")
# min_value=0.0 evita que el usuario ingrese valores negativos
r = st.number_input("Radio del tanque en metros (r):", min_value=0.0)
h = st.number_input("Altura del tanque en metros (h):", min_value=0.0)

# 4. Botón de ejecución y Salida
if st.button("Calcular Volumen"):
    # Proceso
    v = volumen_cilindro(r, h)

    # Salida visual (Reemplaza al print)
    st.success(f"El volumen útil del tanque es de: {v:.2f} metros cúbicos.")
    st.balloons() # ¡Un toque de celebración cuando el cálculo es exitoso!
````
